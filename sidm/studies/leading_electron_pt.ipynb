The overall purpose of this notebook is to replicate figure 1.2 from the SIDM analysis note, which plots the leading lepton pt values in a histogram. This is done for both the 2mu2e and 4mu channels, but the data sets included here do not include the 4mu channel. To do this, four new histograms were created: genE0_pt (leading electron pt from generated events), genE1_pt (subleading electron pt from the generated events), and genMu0_pt and genMu1_pt, which are the analogous plots for muons. Eventually, these plots are placed side-by-side on a logarithmic axis, replicating the figure in the analysis note.

```python
samples = [
    "2Mu2E_100GeV_1p2GeV_9p6mm",
    "2Mu2E_100GeV_1p2GeV_0p096mm",
    "2Mu2E_100GeV_5GeV_4mm",
    "2Mu2E_150GeV_5GeV_2p7mm",
    "2Mu2E_200GeV_5GeV_2mm",
    "2Mu2E_500GeV_5GeV_0p8mm",
    "2Mu2E_800GeV_5GeV_0p5mm",
    "2Mu2E_1000GeV_5GeV_0p4mm",
]
fileset = utilities.make_fileset(samples, "ffntuple_v4")

runner = processor.Runner(
    executor=processor.IterativeExecutor(),
    schema=ffschema.FFSchema,
)

p = sidm_processor.SidmProcessor(
    [
        "2mu2e", "4mu"
    ],
    ["base_plus_gen"]
)

output = runner.run(fileset, treename="ffNtuplizer/ffNtuple", processor_instance=p)
```

```python
hep.histplot(output['out']["2Mu2E_100GeV_1p2GeV_9p6mm"]["hists"]["genE0_pt_lowrange"]["2mu2e", :], label="genE0_pt")
hep.histplot(output['out']["2Mu2E_100GeV_1p2GeV_9p6mm"]["hists"]["genE_pt"]["2mu2e", :100j], label="genE_pt")
hep.histplot(output['out']["2Mu2E_100GeV_1p2GeV_9p6mm"]["hists"]["genE1_pt_lowrange"]["2mu2e", :], label="genE1_pt")
#genE1_pt_lowrange (and mu) was made to show that it and the leading electron / muon line up to trace the start and end of the larger genE histogram.
#This was mostly for testing to show that the leading plot looked right
plt.legend(["leading e", "all e", "subleading e"])
plt.show()
```
